```python
def shot_zone(row):
    if abs(row['x']) < 22 and row['y'] <= 7.8:
        return '2PT'
    elif row['x']**2 + row['y']**2 <= 23.75**2 and row['y']>=7.8:
        return '2PT'
    elif abs(row['x']) >22 and row['y'] < 7.8:
        return 'C3'
    else:
        return 'NC3'
```

```python
df['shotzone'] = df.apply(lambda row: shot_zone(row), axis=1)
```
